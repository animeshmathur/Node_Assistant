```python
training_text = {}
try:
    with open('training_sentences.pkl', mode='rb') as in_file:
        training_text = pickle.load(in_file)
except:
    print('Could not read training_sentences.pkl')
```

```python
training_text.keys()
```

# Insert training text for 'npm list'

```python
if 'npm_list' not in training_text.keys():
    training_text['npm_list'] = ['npm list']
```

```python
input_str = ""
while True:
    input_str = input()
    if input_str.lower() == 'exit':
        break
    if input_str not in training_text['npm_list']:
        training_text['npm_list'].append(input_str)
```

# Store training sentences in file 

```python
with open('training_sentences.pkl', mode='wb') as out_file:
    pickle.dump(training_text, out_file)
```

# Create CSV for training data

```python
training_data = []
for action_key in iter(training_text):
    for text in training_text[action_key]:
        training_data.append({
            "Text": text,
            "Action_Key": action_key
        })
```

```python
training_data
```

```python
training_data_df = pd.DataFrame(training_data)
training_data_df
```

```python
training_data_df = training_data_df[['Text', 'Action_Key']]
```

```python
training_data_df.to_csv('training_data.csv', index=False)
```
